```
import pandas as pd
school = pd.read_csv("http://modcom.co.ke/data2020/datasets/school.csv")
school
```

```
#check empty fields
print(school.isnull().sum())
```

```
#get subset to use
subset1 = school[['Rank', 'Major', 'Gender','Writing', 'Math', 'Reading']]
subset1.tail(100)
```

```
subset1.isnull().sum()
```

```
#Replacing empty field with the a number
subset1['Gender'].fillna(2, inplace=True)
```

```

subset1.tail(100)
```

```
#convert number representation to words
subset1['Gender'].replace({0:'Male', 1:'Female', 2:'Unknown'}, inplace=True)
subset1.tail(100)
```

```
import matplotlib.pyplot as plt
x, y = plt.subplots()
subset1.groupby('Gender').size().plot(kind='pie', autopct="%1.1f%%")
y.set_title("Gender representation")
y.set_xlabel("")
y.set_ylabel("")
```

```
#replacing categogirical missing values
subset1['Rank'].fillna(5, inplace=True)
subset1['Rank'].replace({1:"Freshman", 2:"Sophomore", 3:"Junior", 4:"Senior", 5:"Unknown"}, inplace=True)
subset1['Major'].fillna('Unknown', inplace=True)
subset1
```

```
#fixing and replacing continous data
#replace the value of the missing variable with either a median or a mean value.
median_writing = subset1['Writing'].median()
subset1['Writing'].fillna(median_writing, inplace=True)
#math
median_math = subset1['Math'].median()
subset1['Math'].fillna(median_math, inplace=True)

#English
median_read = subset1['Reading'].median()
subset1['Reading'].fillna(median_read, inplace=True)

subset1
```

```
subset1.isnull().sum()
```

```
import seaborn as sbn
#histogram
x, y = plt.subplots()
y.hist(subset1['Reading'], color="Orange")
y.set_title("Reading Distribution")
y.set_xlabel("Reading out of 100%")
y.set_ylabel("Frequency")
plt.show()
```

```
#Density graph(univariate)
x, y = plt.subplots()
y.plot(subset1['Reading'], color="Orange")
y.set_title("Reading Distribution")
y.set_xlabel("Marks out fo 100")
y.set_ylabel("Number of students")
plt.show()
```

```
#scatter plot graph(bivariate)
x, y = plt.subplots()
y.scatter(subset1['Reading'], subset1['Math'], color="Orange")
y.set_title("Reading vs Math")
y.set_xlabel("Frequency/number of students")
y.set_ylabel("Marks out of 100")
plt.show()
```

```
#stacked bar plot-bivariate or multi variate
subset1.groupby(['Rank', 'Gender'])['Math'].mean().unstack().plot(kind='bar', stacked=False)
y.set_title("Rank by size and gender")
y.set_xlabel("Rank and Gender")
y.set_ylabel("Score by Math")
plt.show()
```

```
#Pie plot bivariate or multi variate
subset1.groupby(['Rank', 'Gender']).size().plot(kind='pie', autopct="%1.1f%%")
y.set_title("Rank by size and gender")
y.set_xlabel("")
y.set_ylabel("")
```
